```python
# Librosa is a Python tool for the analysis of music and audio. 
# Librosa provides the necessary components for developing music information retrieval systems.
import librosa
import librosa.display
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
path='/content/03-01-08-02-01-02-18.wav'
```

```python
def waveplot(data, sr,e):
    plt.figure(figsize=(10, 3))
    plt.title('Waveplot for audio {} emotion'.format(e), size=15)
    librosa.display.waveplot(data, sr=sr)
    plt.show()

def spectrogram(data, sr, e):
    X = librosa.stft(data)
    Xdb = librosa.amplitude_to_db(abs(X))
    plt.figure(figsize=(12, 3))
    plt.title('Spectogram From Audio {} emotion'.format(e), size=15)
    librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')   
    plt.colorbar()
```

```python
emotion='Happy'
Path = np.array(path)
data, sampling_rate = librosa.load(path)
waveplot(data, sampling_rate,emotion)
```
